```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# load netindex dataset
nidf = pd.read_csv('net_index/city_daily_speeds.csv')
nidf.head().T
```

```python
# drop unwanted columns
nidf.drop(columns=['country_code', 'region_code'], inplace=True)
nidf.head().T
```

```python
# group by and calculate mean
nidf2 = nidf.groupby(['country', 'region', 'date']).agg({
    'download_kbps':'mean',
    'upload_kbps':'mean',
    'total_tests':'sum',
    'distance_miles': 'mean'
}).reset_index()
```

```python
# write to file
nidf2.to_csv('net_index/region_daily_speeds.csv', index=False)
```
